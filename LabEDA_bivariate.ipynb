```python
# Import necessary libraries
import pandas as pd
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import skew, kurtosis

# Path to the ZIP file
zip_file_path = 'amz_uk_price_prediction_dataset.csv.zip'

# Open the ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List all files in the ZIP archive
    print("Files in the ZIP archive:", z.namelist())
    
    # Open the CSV file within the ZIP archive
    with z.open('amz_uk_price_prediction_dataset.csv') as csv_file:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(csv_file)
        
df.head()
```

```python
# 1. Crosstab Analysis : product category and isbestseller
crosstab = pd.crosstab(df['category'], df['isBestSeller'], normalize='index')  
```

```python
# 1.1. proportion of best sellers in each category
best_seller =crosstab[True].sort_values(ascending=False)
print(best_seller)
```

```python
# Grocery has the highest prevalence of best sellers, smart home and health and personal care also have a high prevalence with similar levels. 
# In the 4th position we have mobile phones followed by power and hand tools. 
```

```python
# 2. chi_square and crame's V 
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(crosstab)
print(chi2_contingency(crosstab))
```

```python
print(f'Chi2 Statistic: {chi2}, p-value: {p}')
```

H0 = The distribution of the best_sellers is independent of the product category
H1 = The distribution of the best_sellers is not independent of the product category

Chi2 Statistic: 3.9130707743346096
p-value: 1.0 = is high (Common significance levels are 0.05, 0.01, or 0.10.) 

This indicates that there is no statistically significant association between the distribution of best-sellers and the product category
"the best-seller distribution appears to be independent of the product category = H0"

```python
# 2.1
import numpy as np

cramer_v = np.sqrt(chi2 / (len(df) * (min(crosstab.shape) - 1)))
print(f'Cramér\'s V: {cramer_v}')
```

```python
# part 2. Exploring product prices and ratong across categories and brands; 
# 0. How differents product categories influence product prices
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
IQR = q3 - q1
```

```python
dfnout = df[~((df['price'] < (q1 - 1.5 * IQR)) | (df['price'] > (q3 + 1.5 * IQR)))]
```

```python
# visualization violon plots
top_20_categories = dfnout['category'].value_counts().index[:20]
sns.violinplot(x='category', y='price', data=dfnout[dfnout['category'].isin(top_20_categories)])
plt.xticks(rotation=90)
plt.show()
```

```python
# the highest median price = sports & outdoors
```

```python
# 2. bar charts
top_10_categories = dfnout['category'].value_counts().index[:10]
avg_price = dfnout[dfnout['category'].isin(top_10_categories)].groupby('category')['price'].mean()
avg_price.plot(kind='bar')
```

```python
# 3. Box plots
sns.boxplot(x='category', y='stars', data=df[df['category'].isin(top_10_categories)])
plt.xticks(rotation=90)
plt.show()
```

```python
# Which category tends to receive the highest median rating from customers?
Skin cares and birthday gifts have a median around 4.5.  This categories are tighly clustered around the median with fewer outliers.
```

```python
# part 3. Investigatin the interplay between product prices and ratings
# correlation
correlation = dfnout['price'].corr(dfnout['stars'])
print(f'Correlation between Price and Stars: {correlation}')
```

```python
# Is there a significant correlation between product price and its rating? 
# There is a weak negative correlation, that means the price is not a strong predictor of product rating
```
